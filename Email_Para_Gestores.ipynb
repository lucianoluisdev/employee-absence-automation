## Alerta de faltas por Email

```python
import pandas as pd
import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1

# ==========================
# ALERTA
# ==========================
pyautogui.alert("Vai começar. Aperte OK e não mexa em nada.")

# ==========================
# LER BASES
# ==========================
cadastro = pd.read_excel(r'C:\RH_Gestao\Cadastro_Colaboradores.xlsx')
ponto = pd.read_excel(r'C:\RH_Gestao\Base_Faltas_Colaboradores.xlsx')
gestores = pd.read_excel(r'C:\RH_Gestao\Gestores_Setor.xlsx')

# ==========================
# TRATAMENTO
# ==========================
df = pd.merge(cadastro, ponto, left_on='ID', right_on='Id', how='inner')
df = df[['ID', 'Nome', 'Email', 'Setor', 'Faltas']]
df_final = pd.merge(df, gestores, on='Setor', how='left')

# Somente faltosos
df_final = df_final[df_final['Faltas'] > 0]

# ==========================
# ABRIR GMAIL
# ==========================
pyautogui.hotkey('ctrl', 't')
pyautogui.write('mail.google.com')
pyautogui.press('enter')

time.sleep(8)

# ==========================
# ENVIA PARA CADA GESTOR
# ==========================
for i in range(len(gestores)):

    gestor = gestores.loc[i, 'Nome_Gestor']
    email_gestor = gestores.loc[i, 'Email_Gestor']
    setor = gestores.loc[i, 'Setor']

    base_gestor = df_final[df_final['Setor'] == setor]

    # Se não houver faltas no setor
    if len(base_gestor) == 0:
        continue

    lista_faltas = ""

    for j in range(len(base_gestor)):
        colaborador = base_gestor.iloc[j]['Nome']
        faltas = base_gestor.iloc[j]['Faltas']

        lista_faltas += f"{colaborador} - {faltas} falta(s)\n"

    texto = f"""Prezados, bom dia!!!

{gestor},

Segue relatório de faltas do setor {setor}:

{lista_faltas}

Atenciosamente,
RH
"""

    # Botão escrever
    pyautogui.click(x=41, y=174, clicks=1)
    time.sleep(5)
    pyautogui.click(x=1299, y=479, clicks=1)
    time.sleep(3)
    # Digita e-mail
    pyautogui.write(email_gestor)
    pyautogui.press('tab')
    pyautogui.press('tab')

    # Assunto
    pyautogui.write(f'Relatório de Faltas - {setor}')
    pyautogui.press('tab')

    # Corpo
    pyperclip.copy(texto)
    pyautogui.hotkey('ctrl', 'v')

    time.sleep(2)

    # Enviar
    pyautogui.hotkey('ctrl', 'enter')

    time.sleep(5)

print("Todos os e-mails foram enviados.")
```
